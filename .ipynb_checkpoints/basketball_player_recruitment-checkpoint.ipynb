```python
# Dropping several features that are useless in the model making process

dfml.drop(['team_abbreviation', 'college', 'country', 'draft_year', 'draft_round', 'draft_number', 'season'], axis=1, inplace=True)
```

> #### Creating Tuned Machine Learning Models

- Logistic Regression (logreg) Tuned

```python
# Pipeline logreg
# Menggunakan balancing method SMOTE karena target data imbalance
# Menggunakan hyperparameter tuning

smote = SMOTE()
imbpipe_logreg = imbPipeline([('balancing', smote), ('clf', logreg)])
```

```python
imbpipe_logreg.get_params()
```

```python
hyperparam_space_logreg = {
    'clf__C':[100, 10, 1, 0.1, 0.01, 0.001],
    'clf__solver': ['liblinear', 'newton-cg']
}

skfold = StratifiedKFold(n_splits=5, random_state=2021)
gridsearch_logreg = GridSearchCV(
    imbpipe_logreg,
    param_grid = hyperparam_space_logreg,
    cv = skfold,
    scoring = 'precision',
    n_jobs =-1,
    verbose=1
)

gridsearch_logreg.fit(X_train_val, y_train_val)
```

```python
skfold = StratifiedKFold(n_splits = 5, random_state=2021, shuffle=True)
logreg_tuned_cv = cross_val_score(gridsearch_logreg, X_train_val, y_train_val, cv=skfold, scoring='precision')
print('CV logreg Tuned:', logreg_tuned_cv)
print('CV logreg Tuned Mean:', logreg_tuned_cv.mean())
```

__There is a decrease in the precision score of the Tuned Logreg model:__
- CV logreg Tuned Mean: 0.40306534364305274
- CV Logreg Benchmark Mean: 0.6868025084955474

- KNearestNeighbors (KNN) Tuned WITH SMOTE

```python
knn_transformer = ColumnTransformer([
    ('robust', RobustScaler(), list(X.columns))
], remainder = 'passthrough')  

imbpipe_knn = imbPipeline([
    ('transformer', knn_transformer),
    ('balancing', smote),
    ('clf', knn)
])
```

```python
imbpipe_knn.get_params()
```

```python
# KNN Tuned WITH SMOTE

hyperparam_space_knn = {
    'clf__p':[1, 2],
    'clf__weights': ['uniform', 'distance'],
    'clf__n_neighbors': [3, 5, 7, 9, 11]
}

skfold_knn = StratifiedKFold(n_splits=5, random_state=2021)
gridsearch_knn = GridSearchCV(
    imbpipe_knn,
    param_grid = hyperparam_space_knn,
    cv = skfold_knn,
    scoring = 'precision',
    n_jobs =-1,
    verbose=1
)

gridsearch_knn.fit(X_train_val, y_train_val)
```

```python
skfold = StratifiedKFold(n_splits = 5, random_state=2021, shuffle=True)
knn_tuned_cv = cross_val_score(gridsearch_knn, X_train_val, y_train_val, cv=skfold, scoring='precision')
print('CV KNN Tuned WITH SMOTE:', knn_tuned_cv)
print('CV KNN Tuned Mean:', knn_tuned_cv.mean())
```

__There is a decrease in the precision score of the Tuned KNN model with SMOTE:__
- CV KNN Tuned WITH SMOTE Mean: 0.6614578469232122
- CV KNN Benchmark Mean: 0.7550791313949208

- KNearestNeighbors (KNN) Tuned WITHOUT SMOTE

```python
knn_transformer2 = ColumnTransformer([
    ('robust', RobustScaler(), ['Age', 'Height', 'Weight', 'Avg_points', 'Avg_rebounds'])
], remainder = 'passthrough')  

pipe_knn_tuned = Pipeline([
    ('transformer', knn_transformer),
    ('clf', knn)
])
```

```python
pipe_knn_tuned.get_params()
```

```python
hyperparam_space_knn2 = {
    'clf__p':[1, 2],
    'clf__weights': ['uniform', 'distance'],
    'clf__n_neighbors': [3, 5, 7, 9, 11]
}

skfold_knn = StratifiedKFold(n_splits=5, random_state=2021)
gridsearch_knn2 = GridSearchCV(
    pipe_knn_tuned,
    param_grid = hyperparam_space_knn2,
    cv = skfold_knn,
    scoring = 'precision',
    n_jobs =-1,
    verbose=1
)

gridsearch_knn2.fit(X_train_val, y_train_val)
```

```python
skfold = StratifiedKFold(n_splits = 5, random_state=2021, shuffle=True)
knn_tuned_cv2 = cross_val_score(gridsearch_knn2, X_train_val, y_train_val, cv=skfold, scoring='precision')
print('CV KNN Tuned:', knn_tuned_cv2)
print('CV KNN Tuned Mean:', knn_tuned_cv2.mean())
```

__There is an increase in the precision score of the Tuned KNN model without SMOTE:__
- CV KNN Tuned WITHOUT SMOTE Mean: 0.9162716485423056
- CV KNN Benchmark Mean: 0.7550791313949208

__Best KNN Model:__ Tuned KNN WITHOUT SMOTE

- Decision Tree Classifier (DTC) Tuned

```python
smote = SMOTE()
imbpipe_dtc = imbPipeline([('balancing', smote), ('clf', dtc)])
```

```python
imbpipe_dtc.get_params()
```

```python
skfold_dtc = StratifiedKFold(n_splits=5, random_state=2021)

hyperparam_space_dtc = {
    'clf__max_depth': [2,3,4,5,6,7],
    'clf__min_samples_split': [4,5,6,7,8,9,10],
    'clf__criterion': ['entropy', 'gini']
}

gridsearch_dtc = GridSearchCV(
    imbpipe_dtc,
    param_grid = hyperparam_space_dtc,
    cv = skfold_dtc,
    scoring = 'precision',
    n_jobs =-1,
    verbose=1
)

gridsearch_dtc.fit(X_train_val, y_train_val)
```

```python
skfold = StratifiedKFold(n_splits = 5, random_state=2021, shuffle=True)
dtc_tuned_cv = cross_val_score(gridsearch_dtc, X_train_val, y_train_val, cv=skfold, scoring='precision')
print('CV DTC Tuned:', dtc_tuned_cv)
print('CV DTC Tuned Mean:', dtc_tuned_cv.mean())
```

__There is an increase in the precision score of the Tuned DTC model:__
- CV DTC Tuned Mean: 1.0
- CV DTC Benchmark Mean 0.9837057057057057

- #### Best Machine Learning Model for prediction: Tuned DTC Model

```python
dtc_tuned = gridsearch_dtc.best_estimator_.fit(X_train_val, y_train_val)
y_pred_dtc_tuned = dtc_tuned.predict(X_test)

print(classification_report(y_test, y_pred_dtc_tuned))
print('ROC & AUC Score:', roc_auc_score(y_test, y_pred_dtc_tuned))
print('Balanced Accuracy Score:', balanced_accuracy_score(y_test, y_pred_dtc_tuned))
```
